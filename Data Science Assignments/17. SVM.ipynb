```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
```

```python
df['class'] = label_encoder.fit_transform(df['class'])
```

```python
# Encode all categorical features
df_encoded = pd.get_dummies(df)
```

```python
df_encoded
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Define features and target variable
x = df_encoded.drop('class', axis=1)
y = df_encoded['class']
```

```python
# Split the data (80% training, 20% testing)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
sns.pairplot(df,hue='class')
```

```python
sns.countplot(x='class', data=df)
plt.title('Class Distribution')
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='linear')
```

```python
svm_model.fit(x_train, y_train)
```

```python
y_pred = svm_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

# 5. Visualization of SVM Results

```python
# Assuming 2D data (reduce dimensions for visualization if necessary)
from sklearn.decomposition import PCA
```

```python
# Apply PCA for dimensionality reduction (2 components for visualization)
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_test)
```

```python
# Create a scatter plot of the PCA-transformed features
plt.figure(figsize=(10, 6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y_pred, cmap='coolwarm', s=50)
plt.title('SVM Classification Results')
plt.show()
```

# 6. Parameter Tuning and Optimization

```python
# Hyperparameter tuning using GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
# Define the parameter grid
param_grid = {'C': [0.1, 1],
              'gamma': [1, 0.1, 0.01, 0.001],
              'kernel': ['']}
```

```python
# Instantiate the GridSearchCV object
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2)
grid.fit(x_train, y_train)
```

```python
# Best parameters and accuracy
print("Best Parameters:", grid.best_params_)
print("Best Accuracy:", grid.best_score_)
```

```python
# Testing different kernels
kernels = ['linear', 'rbf', 'poly']
for kernel in kernels:
    model = SVC(kernel=kernel)
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    
    print(f"Kernel: {kernel}")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("\n")
```
