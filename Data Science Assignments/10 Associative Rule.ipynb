```python
#Import Libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules,apriori
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_excel("Online retail.xlsx")
```

```python
df
```

```python
df=df.rename(columns={'shrimp,almonds,avocado,vegetables mix,green grapes,whole weat flour,yams,cottage cheese,energy drink,tomato juice,low fat yogurt,green tea,honey,salad,mineral water,salmon,antioxydant juice,frozen smoothie,spinach,olive oil':'products'})
```

```python
df
```

# EDA

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
def txt_split(txt):
    return txt.split(',')
```

```python
data = df['products'].apply(txt_split)
```

```python
data
```

```python
#
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
encoded_df = te.fit_transform(data)
```

```python
encoded_df
```

```python
te.columns_
```

```python
data = pd.DataFrame(encoded_df, columns=te.columns_)
data
```

```python
data.replace([True,False],[1,0], inplace=True)
```

```python
data
```

```python
data.shape
```

```python
scores = apriori(data,min_support = 0.01, use_colnames=True)
scores
```

```python
rules = association_rules(scores,metric="support", min_threshold=0.01)
rules
```

```python
rules.sort_values(by = 'lift', ascending=False)
```
