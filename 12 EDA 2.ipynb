## 4. Feature Selection

```python
from sklearn.ensemble import IsolationForest
```

```python
iso_forest = IsolationForest()
```

```python
outliers = iso_forest.fit_predict(df[numerical_cols])
```

```python
data_no_outliers = df[outliers == 1]
```

```python
#data with no outliers
data_no_outliers
```

```python
#PPS (Predictive Power Score) to Find Relationships Between Features
```

```python
import ppscore as pps 
```

```python
pps_matrix = pps.matrix(data_no_outliers).pivot(columns='x',index='y',values='ppscore')
```

```python
correlation_matrix = data_no_outliers.corr()
```
