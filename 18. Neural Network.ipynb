# 2. Model Implementation

```python
label_encoder = LabelEncoder()
```

```python
df_scaled['letter'] = label_encoder.fit_transform(df_scaled['letter'])
```

```python
x_feature = df_scaled.iloc[:,:-1]
y_feature = df_scaled['letter']
```

```python
xtrain , xtest, ytrain,ytest = train_test_split(x_feature,y_feature,train_size=0.8, random_state=42)
```

## ANN Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(64, input_dim=xtrain.shape[1], activation='relu'))  # Input layer
model.add(Dense(32, activation='relu'))  # Hidden layer
model.add(Dense(len(label_encoder.classes_), activation='softmax'))  # Output layer
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(xtrain, ytrain, epochs=20, batch_size=32, validation_data=(xtest, ytest))
```
